```python
import random as r

account_list = []

class Account:

    now_account = 0 #Q2
    bank = 'SC은행'

    def __init__(self, name, pay):
        self.name = name
        self.pay = int(pay)
        self.depo_his = []
        self.with_his = []

        num1 = r.randint(0,999)
        num2 = r.randint(0,99)
        num3 = r.randint(0,999999)

        self.num01 = str(num1).zfill(3)
        self.num02 = str(num2).zfill(2)
        self.num03 = str(num3).zfill(6)

        self.interest = 0 #Q7
        Account.now_account += 1 #Q2
        account_list.append(self)
        
        

    def get_account_num(self): #Q3
        print(f'생성된 계좌의 수는 {Account.now_account}개 입니다.')

    def deposit(self): #Q4 / Q7
        n = 0
        while n <= 0 :
          try:
              n = int(input("입금 하실 금액을 말씀해주세요. : "))
              if n > 0:
                  self.pay += n
                  m = format(n,',')
                  print("{0}원이 입금되었습니다.".format(m))
                  self.depo_his.append('+'+m+"원")
                  self.interest += 1 #Q7
                  if self.interest >= 5: #Q7
                      x = round(self.pay/100)
                      if x >=1 :
                          print(f"이자발생! 잔액의 1%만큼 이자가 발생했습니다. 이자는 {x}원 입니다.")
                          self.depo_his.append('+ 이자 '+format(x,',')+"원")
                          self.pay += x
                          self.interest = 0
                          print("잔고 :", format(self.pay, ',')+"원")
              else:
                  print("금액을 확인해주십시오.")
          except ValueError:
                print("숫자를 입력해주세요.")      

    def withdraw(self): #Q5
        n = 0
        m = -1
        while m < 0 :
            n = input("출금 하실 금액을 말씀해주세요. : ")
            if n.isdigit() == True :
              n = int(n)
              if n >= 0:
                  if self.pay - n >= 0:
                      m = n
                      self.pay -= m
                      print(f"{m}원이 출금되었습니다. 잔액은 {self.pay}원 입니다.")
                      self.with_his.append('-'+format(m,',')+"원")
                  
                  else:
                      print("잔액을 초과합니다. 출금 금액을 확인해 주십시오.")
              else:
                  print("금액을 확인해주십시오.")   
            else:
              print("숫자를 입력해주십시오.")  

    def display_info(self): #Q6
        print("은행이름 :", self.bank)
        print("에금주 :", self.name)
        print(f"계좌번호 : {self.num01}-{self.num02}-{self.num03}")
        print("잔고 :", format(self.pay, ',')+"원")

    def deposit_history(self):
        print("☆☆ ☆☆☆☆ ☆☆")
        print("☆☆ 입금내역 ☆☆")
        if len(self.depo_his) > 0:
            for i in self.depo_his:
                print(i)

            print("☆☆          ☆☆ ")
            print("☆☆ ☆☆☆☆ ☆☆ ")

        else:
            print("입금내역이 없습니다.")
            print("☆☆          ☆☆ ")
            print("☆☆ ☆☆☆☆ ☆☆ ")
                    
    def withdraw_history(self):
        print("★★ ★★★★ ★★ ")
        print("★★ 출금내역 ★★")
        if len(self.with_his) > 0:
          for i in self.with_his:
              print(i)

          print("★★          ★★ ")
          print("★★ ★★★★ ★★ ")
          
        else:
          print("출금내역이 없습니다.")
          print("★★          ★★ ")
          print("★★ ★★★★ ★★ ")


# def up100():
#   for i in account_list:
#       if i.pay >= 1000000:
#           i.display_info() #Q9
#           print()

# -------------------------------------------------------------------------------------------------------#

a = Account('김영원',100000)
b = Account('김영투',50000)
c = Account('김영쓰리',2000000)

q = 999

while q != 0:
  print("1 : 등록된 이용자 정보")
  print("0 : 프로그램 종료")
  q = input("숫자를 입력하세요.")
  if q.isdigit() == True :
    q = int(q)
    listnum = 1

    if q == 1:
      print("-"*30)
      for i in account_list:
        print(listnum, i.name)
        listnum += 1
      print("-"*30)

      while q != 0:
        q = input('확인하고 싶은 사람의 번호를 입력하십시오.\n0번을 입력하면 프로그램이 종료됩니다.')
        print("-"*30)

        if q.isdigit() == True :
          q = int(q)

          if q in [1,2,3 ]:

            while q != 0:
              user = account_list[q-1]
              print("-"*30)
              print(user.name,"님을 선택하셨습니다.")
              print("원하시는 활동을 선택해주세요.")
              print("1. 정보 확인하기.")
              print("2. 입금하기")
              print("3. 출금하기")
              print("4. 입금내역 확인하기")
              print("5. 출금내역 확인하기")
              print("0. 프로그램 종료하기")
              print("-"*30)
              q1 = input("번호를 입력해주세요.")

              if q1.isdigit() == True :
                q1 = int(q1)

                if q1 == 1:
                  user.display_info()
                  print("-"*30)

                elif q1 == 2:
                  user.deposit()
                  print("-"*30)

                elif q1 == 3:
                  user.withdraw()
                  print("-"*30)

                elif q1 == 4:
                  user.deposit_history()
                  print("-"*30)

                elif q1 == 5:
                  user.withdraw_history()
                  print("-"*30)

                elif q1 == 0:
                  q = 0

              else:
                print("숫자를 확인해주세요")
            else:
              print("입력을 확인해주세요.")
          else:
            print("입력을 확인해주세요.")
        else:
          print("숫자를 입력해주세요")
    else:
      print("입력을 확인해주세요.")
  else:
    print("숫자를 입력해주세요.")
print("*"*30)
print("이용해주셔서 감사합니다.")
print("*"*30)
```
